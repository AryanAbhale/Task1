```python
import re
def extract_numeric_part(text):
    match = re.search(r'\d+', str(text))  # Convert to string before applying regex
    return int(match.group()) if match else None

# Convert values to integers after extracting numeric parts
df['no.of stocks'] = [extract_numeric_part(value) for value in df['number_available_in_stock']]
```
