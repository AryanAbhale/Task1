```python
top5= df['Price in £'].value_counts().head(5)
plt.figure(figsize=(10, 6))
plt.bar(range(1, 6), top5, color='yellow')
plt.xlabel('Rank')
plt.ylabel('Price')
plt.title('Top 5 Counts of Prices')
plt.xticks(range(1, 11))
plt.tight_layout()
```

```python
top_5= df['manufacturer'].value_counts().head(5)

plt.figure(figsize=(10, 6))
plt.bar(range(1,6), top_5)
plt.xlabel('Manufacturer')
plt.ylabel('Number of Products')
plt.title('Number of Products per Manufacturer (Top 5)')
plt.xticks(range(1,6), top_5.index)
plt.tight_layout()

# Show the bar plot
plt.show()
```

```python
df4=df[['product_name','Price in £']]
df4.drop_duplicates()
df4=df.groupby(['product_name'])['Price in £'].sum()
df4=df4.to_frame()
df4=df4.sort_values('Price in £',ascending=False)
df5=df4.head()
df5
```

```python
df5.plot(kind='bar')

plt.title('top 5 expensive products')
plt.xlabel('product_name')
plt.ylabel('Price in £')
plt.xticks(rotation=90)
plt.show()
```

```python
df1=df[['manufacturer','Ratings out of 5 stars']]
df1.drop_duplicates()
df2=df1.groupby(['manufacturer'])['Ratings out of 5 stars'].sum()
df2=df2.to_frame()
df2=df2.sort_values('Ratings out of 5 stars',ascending=False)
df3=df2.head(10)
df3
```

```python
df3.plot(kind='bar')

plt.title('top 10 rated company')
plt.xlabel('manufacturer')
plt.ylabel('Ratings out of 5 stars')
plt.xticks(rotation=90)
plt.show()
```
